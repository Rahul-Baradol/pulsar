```python
import pandas as pd
import matplotlib.pyplot as plt
# Load both datasets
train_df = pd.read_csv("spiral_train.csv")
test_df = pd.read_csv("spiral_test.csv")

# Create subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle("Spiral Train vs Test Data", fontsize=16)

# Plot train data
axes[0].scatter(train_df["x"], train_df["y"], c=train_df["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[0].set_title("Train Data")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

# Plot test data
axes[1].scatter(test_df["x"], test_df["y"], c=test_df["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[1].set_title("Test Data")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
original = pd.read_csv("spiral_test.csv")
predicted = pd.read_csv("spiral_test_prediction_96_64_1_5k.csv")

# Make sure rows align (assumes both files have same x,y order)
correct = (original["label"] == predicted["label"]).sum()
total = len(original)
print(f"Correct predictions: {correct}/{total} ({correct/total:.2%})")

# Create subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle("Results after 5000 steps", fontsize=16)
fig.text(0.5, 0.92, f"Accuracy: {correct/total:.2%}", ha="center", fontsize=12)

# Plot original data
axes[0].scatter(original["x"], original["y"], c=original["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[0].set_title("Original Data")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

# Plot predicted data
axes[1].scatter(predicted["x"], predicted["y"], c=predicted["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[1].set_title("Predicted Data")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()
plt.show()
```
