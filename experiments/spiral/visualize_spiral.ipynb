```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("spiral_data.csv")  # replace with your file path

# Scatter plot
plt.figure(figsize=(6,6))
plt.scatter(
    df["x"], 
    df["y"], 
    c=df["label"], 
    cmap="bwr",   # blue for class 0, red for class 1
    s=20, 
    alpha=0.7
)

plt.title("Spiral Dataset")
plt.xlabel("x")
plt.ylabel("y")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV files
original = pd.read_csv("spiral_test.csv")
predicted = pd.read_csv("spiral_test_prediction_64_64_1_2k.csv")

# Make sure rows align (assumes both files have same x,y order)
correct = (original["label"] == predicted["label"]).sum()
total = len(original)
print(f"Correct predictions: {correct}/{total} ({correct/total:.2%})")

# Create subplots for side-by-side comparison
fig, axes = plt.subplots(1, 2, figsize=(12, 6))
fig.suptitle("Results after 2000 steps", fontsize=16)
fig.text(0.5, 0.92, f"Accuracy: {correct/total:.2%}", ha="center", fontsize=12)

# Plot original data
axes[0].scatter(original["x"], original["y"], c=original["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[0].set_title("Original Data")
axes[0].set_xlabel("x")
axes[0].set_ylabel("y")

# Plot predicted data
axes[1].scatter(predicted["x"], predicted["y"], c=predicted["label"], cmap="bwr", alpha=0.7, edgecolors="k")
axes[1].set_title("Predicted Data")
axes[1].set_xlabel("x")
axes[1].set_ylabel("y")

plt.tight_layout()
plt.show()
```
